```python
import requests
import json

url_generate = "http://localhost:11434/api/generate"

payload = {
    "model": "phi3",
    "prompt": "Explain how transformers work in 3 sentences.",
    "stream": False
}

response = requests.post(url_generate, json=payload)

print(response.json()["response"])
```

```python
url_chat = "http://localhost:11434/api/chat"

payload = {
    "model": "phi3",
    "messages":[
        {"role": "system", "content": "Answer the user question with a short message with all capital letters"},
        {"role": "user", "content": "What does a strawberry taste like?"}
    ],
    "stream": False
}

response = requests.post(url_chat, json=payload)

print(response.json()["message"]["content"])
```
